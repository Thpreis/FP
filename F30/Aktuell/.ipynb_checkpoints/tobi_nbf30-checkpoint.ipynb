```python
#4.2
k = 8.617e-5
x=np.linspace(158, 275, 1000)
x2=np.linspace(3.5e-3, 5.25e-3, 1000)
def Int1(t, c, Eg):
    return c*t**(3/2)*exp(-Eg/(2*k*t))
def Int2(t, c, Eg):
    return c*exp(-Eg*t/(2*k))

plt.figure()
plt.errorbar(T, I, yerr=sqrt(abs(I)), fmt='.')
popt, pcov = curve_fit(Int1, T, I, sigma=sqrt(abs(I)), p0=[7e11, 1.18])

plt.plot(x,Int1(x, *popt),color="red")
plt.title('Temperature dependency of the dark current')
plt.xlabel('T [K]')
plt.ylabel('dark current [counts]')
plt.savefig('diag1.pdf', format='PDF')

plt.figure()
plt.errorbar(1/T*1e3, I*T**(-3/2), yerr=sqrt(abs(I))*T**(-3/2), fmt='.')
popt2, pcov2 = curve_fit(Int2, 1/T, I*T**(-3/2), sigma=sqrt(abs(I))*T**(-3/2), p0=[1e10, 1.2])
plt.plot(x2*1e3,Int2(x2, *popt2))

Eg = round(popt2[1],3)
dEg= round(sqrt(pcov2[1,1]), 3)
co = popt2[0]

plt.annotate(r'$y={} \cdot 10^{{{}}} \cdot exp(-{}x)$'.format(int(round(co/10**floor(log10(co)), 2)), int(floor(log10(co))), int(floor(round(Eg/2/k)))), (4.4, 1), color='orange')
plt.annotate('$E_g = ({} \pm {})eV$'.format(Eg, dEg), (4.4,0.5))
plt.yscale('log')
plt.xlim(3.5, 5.25)
plt.ylim(1e-4, 15)
plt.title('Determination of the band gap')
plt.xlabel('$mT^{-1}$[$K^{-1}$]')
plt.ylabel(r'$I \cdot T^{-3/2}$ [$counts \cdot K^{-3/2}$]')
plt.savefig('diag2.pdf', format='PDF')


print('Eg = ({}±{})eV'.format(Eg, dEg))
print('med-Bias start: {} ± {}'.format(int(round(Bias[0])), int(round(dBias[0]))))
print('med-Bias end: {} ± {}'.format(round(Bias[-1], 1), round(dBias[-1], 1)))
```

```python
#We average over the pairs of images in order to obtain one value per exposure time
s_tot_d2 = (s_tot_d21+s_tot_d22)/2

#4.5.3.1
s_r_d2=(s_r_d21+s_r_d22)/2
s_r_d1=s_r_d2[0]#Readout noise 


#Two images with the same signal level, thus exposure time, should have the same PRNU noise
#We can thus compute the PRNU noise via the difference noise and the total noise measured in the same area
s_PRNU_e1=np.abs(s_tot_d2[0]-s_diff_d2[0])#PRNU noise

#Totalnoise 
s_tot_d1 = s_tot_d2[0]
#Photonnoise via equation (21)
s_e_d1 =np.sqrt(np.abs(s_tot_d1**2-s_PRNU_e1**2-s_r_d1**2))

print("Photonnoise",s_e_d1)
print("Readoutnoise",s_r_d1)
print("PRNU noise",s_PRNU_e1)
#now explain why the photonnoise dominates !
```

```python
#Does the PRNU NOise dominate for high signal values ? This is for the highstes available exposure time
#We observe here that PRNU noise rises strongly depending on the exposure time
s_r_d10=s_r_d2[9]#Readout noise 


#Two images with the same signal level, thus exposure time, should have the same PRNU noise
#We can thus compute the PRNU noise via the difference noise and the total noise measured in the same area
s_PRNU_e10=np.abs(s_tot_d2[9]-s_diff_d2[9]) #PRNU noise

#Totalnoise 
s_tot_d10 = s_tot_d2[9]
#Photonnoise via equation (21)
s_e_d10 =sqrt(np.abs(s_tot_d10**2-s_PRNU_e1**2-s_r_d10**2))

print("Photonnoise",s_e_d10)
print("Readoutnoise",s_r_d10)
print("PRNU noise",s_PRNU_e10)
#now explain why the photonnoise dominates !
```

```python
PRNUfactor1 = s_PRNU_e10/10000
PRNUfactor2 = s_e_d10/10000
print(PRNUfactor1)
print(PRNUfactor2)
```

```python
#4.5.3.3 Calculate gain via (19)
#Readoutnoise
s_r_d=s_r_d2
#totalnoise
s_tot_d=s_tot_d2
#PRNUnoise
s_PRNU_e=np.abs(s_tot_d2-s_diff_d2)
#Differencenoise
s_diff_d=s_diff_d2
#Photoncount
N_e_d = (Ne_d1+Ne_d2)/2
#Photonnoise

s_e_d =sqrt(np.abs(s_tot_d**2-s_PRNU_e**2-s_r_d**2))
#Gain values for different exposure times via (19) with eta=1 for consistency
kappa2=N_e_d/s_e_d**2
print(kappa2)
print("kappa for one image with ~ 30000 counts","=", kappa2[9], "+-", np.std(kappa2))
```

```python
#4.5.3.2 
#rearranging terms we want to plot and fit the function following from (23):
#(s_diff_d**2/2-s_r_d**2) = 1/kappa * N_e_d 
def fit(x,A):
    return A*x
yvalues=np.abs((s_diff_d**2)/2-s_r_d**2)
xvalues=N_e_d
popt,pcov=curve_fit(fit, xvalues, yvalues)
plt.plot(xvalues,fit(xvalues,*popt))
plt.scatter(xvalues,yvalues)
plt.title('Sensitivity of the detector')
plt.xlabel('$N_{e,d}$ [counts]')
plt.ylabel('$\sigma^2_{diff,d}/2 - \sigma^2_{Read,d}$')

plt.savefig('sensitivity.pdf', format='PDF')
kappa1=1/popt[0]
dkappa1=(pcov[0,0]**0.5)
print("Kappa via equation (23)","=",kappa1, "+-",dkappa1)
```

```python
l = np.abs(kappa1-kappa2[9])/np.sqrt(dkappa1**2+np.std(kappa2)**2)
print("Discrepancy between electron sensitivities", l)
```
